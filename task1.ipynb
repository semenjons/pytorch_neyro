```python
import torch
```

```python
x = torch.randn(3,4,5) .type(torch.FloatTensor)  # Создайте случайный FloatTensor размера 3x4x5
print(x)
```

```python
x.shape #Выведите его форму (shape)
```

```python
z = x.view(6, 10)  # Приведите его к форме 6х10
z
```

```python
v = torch.randperm(10) #Создайте случайный IntTensor размера 10
v
```

```python
r = torch.mul(z,v) #Умножьте созданный FloatTensor на созданный IntTensor поэлементно
r
```

```python
# Умножьте тензор матрично на себя, чтобы результат был размерности 6x6 Подсказка: .dot может не подойти

g = torch.mul(r,r)
g
```

```python
#  Посчитайте производную функции y = x**3 + z - 75t в точке (1, 0.5, 2)

import torch

x = torch.tensor(1.0, requires_grad=True)
z = torch.tensor(0.5, requires_grad=True)
t = torch.tensor(2.0, requires_grad=True)

y = x**3 + z - 75*t

y.backward()

print('Производная по x:', x.grad)
print('Производная по z:', z.grad)
print('Производная по t:', t.grad)
```

```python
# Создайте единичный тензор размера 5x6
tensor = torch.ones(5, 6)
print(tensor)
```

```python
# Переведите его в формат numpy
import numpy as np

numpy_array = tensor.numpy()
print(numpy_array)
```
